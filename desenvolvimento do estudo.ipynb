```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib.patches import Patch
import matplotlib.ticker as mticker

# Carregar os dados do arquivo CSV para um DataFrame
df = pd.read_csv('trabalho_cont.csv', delimiter=';')

# Dicionário de mapeamento de número de mês para nome do mês
meses = {5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto', 9: 'Setembro'}

# Substituir os números dos meses pelos nomes dos meses na coluna 'MES'
df['MES'] = df['MES'].map(meses)

# Remover linhas com valores NaN, se houver
df.dropna(inplace=True)

# Reordenar o DataFrame de acordo com a ordem dos meses
mes_order = ['Maio', 'Junho', 'Julho', 'Agosto', 'Setembro']
df['MES'] = pd.Categorical(df['MES'], categories=mes_order, ordered=True)
df_sorted = df.sort_values('MES')

# Exibir as primeiras linhas do DataFrame para garantir que os dados foram carregados corretamente
print(df.head())
```

```python
# Definindo a largura das barras
bar_width = 0.25

# Criando o intervalo para as posições das barras
x = np.arange(len(df_sorted))

# Plotar as barras para cada categoria
plt.figure(figsize=(12, 6))
plt.bar(x - bar_width, df_sorted['Trabalhando'], width=bar_width, label='Trabalhando')
plt.bar(x, df_sorted['Nao_trabalhando'], width=bar_width, label='Não Trabalhando')
plt.bar(x + bar_width, df_sorted['mais_de_um_emprego'], width=bar_width, label='Mais de um Emprego')

# Definindo os rótulos dos ticks no eixo x
plt.xticks(x, df_sorted['MES'])

plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.title('Distribuição de Status de Emprego por Mês')
plt.legend()
plt.grid(True)
plt.show()
```
